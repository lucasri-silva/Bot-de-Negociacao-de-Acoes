```python
import yfinance as yf
import pandas as pd
import investpy as inv
import os
```

```python
br = inv.stocks.get_stocks(country='brazil')
```

```python
br
```

```python
carteira = []

for ativo in br['symbol']:
    if(ativo == 'ABEV3'):
        carteira.append(ativo+'.SA')
```

```python
carteira
```

```python
## download dados com intervalo 1d entre as datas estabelecidas
dt = yf.download(carteira, start='2016-01-01', end='2023-01-01')
```

```python
dt
```

```python
os.makedirs('./dataset/', exist_ok=True)
dt.to_excel('./dataset/ABEV3_diario.xlsx', index=True) #save downloaded dataframe
```
